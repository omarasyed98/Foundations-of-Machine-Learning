```python
#Problem 2/3 Object Oriented
import numpy
import matplotlib.pyplot as plt

class Buffon:
    
    def __init__(self,l,t,n,m):
        self.l = l
        self.t = t
        self.n = n
        self.m = m
        self.x = []
        self.theta = []
        self.Prob_Graph = []
        self.Rat = []
        self.Analytic = []
    
    def random(self):
        for i in range(self.n):
            self.x.append(numpy.random.uniform(0, self.t/2))
            self.theta.append(numpy.random.uniform(0,numpy.pi/2))
        return [self.x,self.theta]
    
    def simulation(self):
        self.x=[]
        self.theta=[]
        self.random()
        a=0
        for i in range(self.n):
            if self.x[i] < self.l/2*numpy.sin(self.theta[i]):
                a+=1
            else:
                continue
        return (a/self.n)
    
    def pi(self):
        return(self.simulation()*self.t*10)/(2*self.l)
        


    def Prob(self):
        b=0
        for i in range(self.m):
            self.Prob_Graph.append(Buffon(b,1,1000,self.m).simulation())
            b+=0.1
        return self.Prob_Graph
    
    def Ratio(self):
        c=0
        for i in range(self.m):
            self.Rat.append(c)
            c+=0.1
        return self.Rat
    
    def Anal(self):
        return(2*self.l)/(self.t*numpy.pi)
    
    def Anal1(self):
        r=self.l/self.t
        return(2/numpy.pi)*((r-numpy.sqrt(r**2-1))+(numpy.arccos(1/r)))
    
    def Analytic_Sol(self):
        d=0
        for i in range (self.m):
            if d<=1:
                self.Analytic.append(Buffon(d,1,1000,50).Anal())
                d+=0.1
            else:
                self.Analytic.append(Buffon(d,1,1000,50).Anal1())
                d+=0.1
        return self.Analytic
        
print("Probabilty =", Buffon(2,3,1000,50).simulation())
print("Pi =", Buffon(2,3,1000,50).pi())
    
plt.plot(Buffon(2,3,1000,50).Ratio(),Buffon(2,3,1000,50).Prob())
plt.title (r"Probabilty vs l/t")
plt.xlabel(r"Ratio of l/t")
plt.ylabel(r"Probability")




    



        
            
        
```

```python
plt.plot(Buffon(2,3,1000,50).Ratio(),Buffon(2,3,1000,50).Prob(), 'r-', label='Numerical Solution')
plt.plot(Buffon(2,3,1000,50).Ratio(),Buffon(2,3,1000,50).Analytic_Sol(), 'y-', label='Analytical Solution')
plt.legend()
plt.title (r"Probabilty vs l/t")
plt.xlabel(r"Ratio of l/t")
plt.ylabel(r"Probability")
```

```python
def Fib_rec (n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return Fib_rec(n-1)+Fib_rec(n-2)
    
#print(Fib_rec(10))

import time
def timer(n, k, f=Fib_rec):
    x=time.time()
    for i in range(k):
        a=f(n)
    y=time.time()
    average_time = (y-x)/k
    return(average_time)

recursive_list=[]
def list (b):
    for i in range(35):
        recursive_list.append(timer(b,5))
        b=b+1
list(0)

import matplotlib.pyplot as plt
x = [i for i in range(0,35)]
y= recursive_list

plt.plot(x,y)
plt.title (r"Recursive vs. Iterative Fibonacci Sequence")
plt.xlabel(r"$Position in Fibonacci Sequence$")
plt.ylabel(r"$Average Time to Compute$")


```

```python
#Problem #2/3
#low number of simulations taken to see significant difference between analytical and numerical solution
import numpy
import matplotlib.pyplot

def general (t,l):
    a=0
    b=0
    x= numpy.random.uniform (0,t/2,1000)
    theta= numpy.random.uniform (0,numpy.pi/2,1000)
    for i in range(len(x)):
        if x[i] < l/2*numpy.sin(theta[i]):
            a+=1
        else:
            b+=1
    return (a/(a+b))
    #print("Probability =", a/(a+b))


print("Probability=", general(3,2))

print("Pi=", (general(3,2)*3*10)/(2*2))
```

```python
#Problem #4
import matplotlib.pyplot as plt

Probability=[]
def Prob(n):
    b=0
    for i in range(n):
        Probability.append(general(1,b))
        b+=0.1
    return Probability

y=Prob(50)

Ratio=[]
def Rat(n):
    c=0
    for i in range (n):
        Ratio.append(c)
        c+=0.1
    return Ratio

x=Rat(50)


plt.plot(x,y)
plt.title (r"Probabilty vs l/t")
plt.xlabel(r"Ratio of l/t")
plt.ylabel(r"Probability")

```

```python
#Problem #5
import numpy
import matplotlib.pyplot as plt

def Anal(t,l):
    return(2*l)/(t*numpy.pi)


def Anal1 (t,l):
    r=l/t
    return(2/numpy.pi)*((r-numpy.sqrt(r**2-1))+(numpy.arccos(1/r)))

Analytic=[]
def Analytic_Sol(n):
    d=0
    for i in range (n):
        if d<=1:
            Analytic.append(Anal(1,d))
            d+=0.1
        else:
            Analytic.append(Anal1(1,d))
            d+=0.1
    return Analytic

z=Analytic_Sol(50)


plt.plot(x,y, 'r-', label='Numerical Solution')
plt.plot(x,z, 'y-', label='Analytical Solution')
plt.legend()
plt.title (r"Probabilty vs l/t")
plt.xlabel(r"Ratio of l/t")
plt.ylabel(r"Probability")

#When the number of simulations is increased past 1,000 there is little difference between the Numerical and Analytical solutions for the Probability


    
```
